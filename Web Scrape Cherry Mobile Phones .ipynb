# Get the Model Name and Price using Regex in Python

```python
#Packages
import urllib
import re

url='http://www.pinoytechnoguide.com/2013/01/cherry-mobile-price-list.html' #website

phone_name=re.compile('<h2>(.+?)</h2>') #set the pattern
    
    
htmlfile=urllib.urlopen(url)
htmltext = htmlfile.read()
phone_list=re.findall(phone_name,htmltext) #find the data inside the pattern
   
print phone_list

    
```

# Clean the Data

```python
Phone_price=[]
for i in phone_list:
    out = re.sub("&#8369;", '', i)
    out1 = re.sub("</strike>", '', out)
    out2 = re.sub("<strike>", '', out1)
    Phone_price.append(out2)

for x in Phone_price[1:]:
    print x
```

# Separate Phone Names to Prices

```python
Sep_phone_price=[]
for line in Phone_price[1:]:
    Sep_phone_price.append([word for word in line.split("-") if word])
```
